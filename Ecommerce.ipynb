```python
import pandas as pd
df = pd.read_excel("/content/drive/MyDrive/Ecommerce.xlsx")
df.head()
```

```python
df.info() # check data
```

```python
df = df.dropna() # remove missing values
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df ['event_time'] = pd.to_datetime(df['event_time'])  # convert time
```

```python
df.dtypes
```

```python
df = df[df['price'] > 0] # filtered important data
```

```python
df['purchase'] = df['event_type'].apply(lambda x:1 if x == 'purchase' else 0) # create purchase column
```

```python
df['hour'] = df['event_time'].dt.hour
df['day'] = df['event_time'].dt.day
```

```python
user_data = df.groupby('user_id').agg({
    'event_type': 'count',   # activity level
    'price': 'mean',
    'purchase': 'max'
}).reset_index()

user_data.columns = ['user_id', 'activity', 'avg_price', 'purchase']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X = user_data[['activity', 'avg_price']]
y = user_data['purchase']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, y_train)

accuracy = model.score(X_test, y_test)
print("Accuracy:", accuracy)
```
